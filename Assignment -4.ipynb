# 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
import pandas as pd

from selenium import webdriver
from bs4 import BeautifulSoup
import numpy as np
import time
import re
import requests
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import WebDriverException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By


import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# Opening Wikipedia webpage
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
time.sleep(2)
```

```python
# Empty list for scraping data
Rank =[]
Name =[]
Artist =[]
Upload_date=[]
Views=[]
```

```python
# Extracting Rank Via X path
try:
    rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rank:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('NA')
except StaleElementReferenceException:
    Rank.append('NA')
    
# Extracting Name Via X path
try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in name:
        Name.append(i.text)
except NoSuchElementExceptionhElementException:
    Name.append('NA')
except StaleElementReferenceException:
    Name.append('NA')
    
# Extracting Artist Name Via Xpath
try:
    artist=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist:
        Artist.append(i.text)
except NoSuchElementExceptionhElementException:
    Artist.append('NA')
except StaleElementReferenceException:
    Artist.append('NA')

# Extracting Upload date Via Xpath
try:
    upload_date=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_date:
        Upload_date.append(i.text)
except NoSuchElementExceptionhElementException:
    Upload_date.append('NA')
except StaleElementReferenceException:
    Upload_date.append('NA') 

# Extracting Views via Xpath
try:
    views=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views:
        Views.append(i.text)
except NoSuchElementExceptionhElementException:
    Views.append('NA')
except StaleElementReferenceException:
    Views.append('NA')
```

```python
# Dataframe for scrap data
Wiki_YT=pd.DataFrame({'Rank':Rank,'Video Name':Name,'Uploader':Artist,'Views (in Billons)':Views,'Upload Date':Upload_date})
print('\033[1m'+'Most Viewed Video on YouTube from Wikipedia :'+'\033[0m')
Wiki_YT
```

```python
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
votes=[]
```

```python
for i in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/h3/a"):
    name.append(i.text)
#year_span
for year in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/h3/span[2]"):
    year_span.append(year.text)
#genre
for g in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/p/span[5]"):
    genre.append(g.text)
#run_time
for time in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/p/span[3]"):
    run_time.append(time.text)
#ratings
for v in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/p[4]/span[2]"):
    votes.append(v.text)
#votes
for rate in driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/div[1]/div/span[2]"):
    ratings.append(rate.text)
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Name"]=name
df["Year span"]=year_span
df["Genre"]=genre
df["Run time"]=run_time
df["Ratings"]=ratings
df["Votes"]=votes
df
```
